```pyspark
# from __future__ import print_function

import sys

import numpy as np
from pyspark.sql import SparkSession
from datetime import datetime, timedelta
import pyspark.sql.functions as F
import json
from pyspark.sql import types as T
import pytz
import hashlib
import pytz
import re
patten='("phone"|"phone_number"|"user_mobile"|"number"|"id_card_no"|"nik"|"caller"|"callee"|"to"):"(?:[^"\\\\]|\\\\.)*"'

def match_replaceQuote(matchObj):
    dic=matchObj.group()
    diclist=dic.split(":")
    str1=str(diclist[0])+':"'+str(hashlib.sha256(str(diclist[1].replace('"',"")).encode('utf-8')).hexdigest() if diclist[1]!=None else diclist[1])+'"'
    return str1
def parseJson(d):
    if d != None:
        return re.compile(patten).sub(match_replaceQuote,d)
    else:
        return d
spark.udf.register("parseJson",parseJson)
spark.sql("""
 select thirdparty_id, update_time, retry_times, create_time, parseJson(regexp_replace(response, '(\d*)(8|62)(\d{4})(\d*)', '$1$2xxxx$4')) response,channel,id,customer_id,loan_id,status 
 from  from  thirdparty_etl.thirdparty_t_task  limit 10000
""").show(100,False)
```
